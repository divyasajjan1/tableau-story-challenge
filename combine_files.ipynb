```python
import pandas as pd
import os
import random

# Function to list all CSV files in a directory
def list_csv_files(directory):
    csv_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".csv"):
                csv_files.append(os.path.join(root, file))
    return csv_files

# Generate file paths for each month's data in 2022, and 2023 
file_paths = [
    list_csv_files("2022-citibike-tripdata/1_January"),  # List of CSV files for January 2022
    list_csv_files("2022-citibike-tripdata/2_February"),  # List of CSV files for February 2022
    list_csv_files("2022-citibike-tripdata/3_March"),  # List of CSV files for March 2022
    list_csv_files("2022-citibike-tripdata/4_April"),  # List of CSV files for April 2022
    list_csv_files("2022-citibike-tripdata/5_May"),  # List of CSV files for May 2022
    list_csv_files("2022-citibike-tripdata/6_June"),  # List of CSV files for June 2022
    list_csv_files("2022-citibike-tripdata/7_July"),  # List of CSV files for July 2022
    list_csv_files("2022-citibike-tripdata/8_August"),  # List of CSV files for August 2022
    list_csv_files("2022-citibike-tripdata/9_September"),  # List of CSV files for September 2022
    list_csv_files("2022-citibike-tripdata/10_October"),  # List of CSV files for October 2022
    list_csv_files("2022-citibike-tripdata/11_November"),  # List of CSV files for November 2022
    list_csv_files("2022-citibike-tripdata/12_December"),  # List of CSV files for December 2022
    list_csv_files("2023-citibike-tripdata/1_January"),  # List of CSV files for January 2023
    list_csv_files("2023-citibike-tripdata/2_February"),  # List of CSV files for February 2023
    list_csv_files("2023-citibike-tripdata/3_March"),  # List of CSV files for March 2023
    list_csv_files("2023-citibike-tripdata/4_April"),  # List of CSV files for April 2023
    list_csv_files("2023-citibike-tripdata/5_May"),  # List of CSV files for May 2023
    list_csv_files("2023-citibike-tripdata/6_June"),  # List of CSV files for June 2023
    list_csv_files("2023-citibike-tripdata/7_July"),  # List of CSV files for July 2023
    list_csv_files("2023-citibike-tripdata/8_August"),  # List of CSV files for August 2023
    list_csv_files("2023-citibike-tripdata/9_September"),  # List of CSV files for September 2023
    list_csv_files("2023-citibike-tripdata/10_October"),  # List of CSV files for October 2023
    list_csv_files("2023-citibike-tripdata/11_November"),  # List of CSV files for November 2023
    list_csv_files("2023-citibike-tripdata/12_December"),  # List of CSV files for December 2023
]

# Initialize an empty list to store sampled data
sampled_data = []

# Sample 1% of rows from each file
for month_files in file_paths:
    month_sampled_data = []
    for file in month_files:
        # Read the CSV file
        df = pd.read_csv(file, low_memory=False)
        # Calculate the number of rows to sample (1% of total rows)
        sample_size = max(1, int(len(df) * 0.05))
        # Sample rows randomly
        sampled_rows = df.sample(n=sample_size, random_state=42)
        # Append sampled rows to the list
        month_sampled_data.append(sampled_rows)
    # Concatenate sampled data for each month
    sampled_data.append(pd.concat(month_sampled_data, ignore_index=True))

# Concatenate sampled data from all months
final_result = pd.concat(sampled_data, ignore_index=True)

# Display the result
final_result
```

```python
# Convert 'started_at' column to datetime
final_result['started_at'] = pd.to_datetime(final_result['started_at'])

# Extract year and month from 'started_at' column
final_result['year_month'] = final_result['started_at'].dt.to_period('M')

# Count the number of samples by year and month
year_month_counts = final_result['year_month'].value_counts()

# Sort the counts by year and month
year_month_counts_sorted = year_month_counts.sort_index()

# Display the sorted counts
year_month_counts_sorted
```

```python
#Making a copy for any further checks on the dataframe so that the dataframe does not get disturbed
result = final_result.copy()
result
```

```python
import numpy as np

# Define a function to calculate the Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 3956  # Radius of Earth in miles. Use 6371 for kms
    return c * r

# Apply the function to calculate the distance and store it in a new column
result['distance_miles'] = haversine(result['start_lat'], result['start_lng'], result['end_lat'], result['end_lng']).round(2)
result
```
